```python
#Imports
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
# Make graphs like the Sat-SINR ablations that show the scaling of the four combinations ()

def plotot(file_name, y_label, sizes, vals_base, vals_rc, vals_chl, vals_rc_chl):
    fig, ax = plt.subplots(1)
    
    ax.plot(sizes, vals_base, "--o", label='Siren', linewidth=1, color="blue")
    ax.plot(sizes, vals_rc, "--+", label='ReSiren', linewidth=1, color="green")
    ax.plot(sizes,vals_chl, "--^", label='Siren + ChL', linewidth=1, color="brown")
    ax.plot(sizes, vals_rc_chl, "--*", label='ReSiren + ChL', linewidth=1, color="red")
    #ax.tick_params(axis='y', labelcolor=f1s_col)
    ax.set_xlabel('Number of layers')
    ax.set_ylabel(y_label)
    #ax.set_ylim([7.01,7.55])
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, loc="upper left")
    
    #plt.savefig(file_name)
```

```python
sizes = ["2", "4", "8", "16", "32"]
vals_base = [86.9281,88.75561,89.0662,88.55944,85.00000]
vals_rc = [85.00000,85.00000,85.00000,85.00000,85.00000]
vals_chl = [87.06768,89.20237,89.27687,89.10379,89.17447]
vals_rc_chl = [85.00000,85.00000,85.00000,85.00000,85.00000]
plotot("biomes_weighted_f1", "Biomes classification weighted F-1", sizes, vals_base, vals_rc, vals_chl, vals_rc_chl)
```
