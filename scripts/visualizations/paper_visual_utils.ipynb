```python
#Imports
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
# Make graphs like the Sat-SINR ablations that show the scaling of the four combinations ()


font = {'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)

def plotot(file_name, y_label, sizes, vals_base, vals_rc, vals_chl, vals_rc_chl, pos="lower left"):
    fig, ax = plt.subplots(1)

    if vals_base[0] < 1:
        vals_base = [100 * v for v in vals_base]
        vals_rc = [100 * v for v in vals_rc]
        vals_chl = [100 * v for v in vals_chl]
        vals_rc_chl = [100 * v for v in vals_rc_chl]
    
    #ax.plot(sizes, vals_base, "--o", label='SIREN', linewidth=1, color="blue")
    ax.plot(sizes, vals_rc, "--^", label='ReSIREN', linewidth=1, color="green")
    ax.plot(sizes[:4],vals_chl, "--^", label='ResFFN ReLU', linewidth=1, color="brown")
    ax.plot(sizes[:4], vals_rc_chl, "--*", label='ResFFN GELU', linewidth=1, color="red")
    #ax.tick_params(axis='y', labelcolor=f1s_col)
    ax.set_xlabel('Number of layers')
    ax.set_ylabel(y_label)
    #ax.set_ylim([7.01,7.55])
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, loc=pos)
    
    plt.savefig(file_name)
```

```python
sizes = ["2", "4", "8", "16", "32"]
vals_base = [88.4939,90.3526,90.1330,90.0326,90.4347]
vals_rc = [87.118,90.0769,90.7845,90.9397,90.5541]
vals_chl = [87.06768,89.20237,89.27687,74.5785,89.17447]
vals_rc_chl = [85.97676,88.90765,90.07546,76.7306,89.13521]
plotot("biomes_weighted_f1", "Biomes classification weighted F1 in %", sizes, vals_base, vals_rc, vals_chl, vals_rc_chl)
```
